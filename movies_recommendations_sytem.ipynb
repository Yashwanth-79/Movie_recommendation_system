# Movie Recommendation System


Movie recommendation systems are software tools that suggest movies to users based on their preferences, using machine learning algorithms and data about user behavior and movie attributes.

> Types of recommendation system
  1. Content Based recommendation system 
  2. Popularity Based recommendation system 
  3. Collaborative Recommendation System

##### Some platforms that uses "Movie Recommendation System" are Netflix,Amazon prime, Disney Hotstar,etc

## Let's Begin

This project workflow : 

Data --> Data preprocessing --> feature extraction --> User Input --> Cosine Similarity Algorithm --> List Of Movies

Importing the dependencies

```python
import numpy as np 
import pandas as pd 
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre processing 

```python
# loading the data from csv file to dataframe 
movies_data = pd.read_csv('movies.csv')
# printing 5 rows of data frame
movies_data.head()
```

```python
# no. of rows and columns 
movies_data.shape
```

```python
movies_data.isna().sum()
```

```python
# selecting the relevant features for recommendation
selected_features = ['genres','keywords','tagline','cast','director','original_title']
selected_features
```

```python
# replacing the null values with null string 
for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna(' ')
```

```python
# combining all the selected feature 
combined_feature = movies_data.genres + ' ' + movies_data.keywords + ' ' + movies_data.tagline + ' ' + movies_data.cast + ' ' +movies_data.director + ' ' + movies_data.original_title
combined_feature.head()
```

```python
# converting text data into feature vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_feature)
print(feature_vectors)
```

Cosine Similarity 


```python
# getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

Taking input from user 


```python
#getting the movie name from the user 
movie_name = input('Enter your interested English movie name : ')
print(movie_name)
```

```python
#creating a list with all the movies given in the dataset
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the closest match for given movie by user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print( index_of_the_movie )
```

```python
#getting the list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1] ,reverse=True)
print(sorted_similar_movies)
```

```python
# printing name of similar movies based on index
print('Suggested movies for you : \n')

i =1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if i<11:
        print(i , '.',title_from_index)
        i+=1
```
